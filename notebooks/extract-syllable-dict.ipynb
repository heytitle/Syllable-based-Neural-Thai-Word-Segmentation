```python
import sys
sys.path.append("../") 
import attacut
```

```python
DATA_PATH = "../data/best-raw/training.txt"
```

```python
def split_words_and_syllable_segment(line):
    words = line.split("|")
    syllables = []
    for w in words:
        syllables.extend(attacut.preprocessing.syllable_tokenize(w))
    return syllables
split_words_and_syllable_segment("บ้าน|แม่|...|ที่|จังหวัดพิษณุโลก|นั้น| |อะไร|ๆ| |ก็|ดี|ไป|เสีย|หมด| |แม่|เป็น|คน|ที่| จังหวัด|นิยม| |อย่าง|รุนแรง|จนกระทั่ง|ถูก|น้ารัน")
```

```python
count_syllables = dict()

with open(DATA_PATH, "r") as fh:
    for i, line in enumerate(fh):
        if (i+1) % 1000 == 0:
            print("completed", i)
        syllables = split_words_and_syllable_segment(line.strip())

        for sy in syllables:
            if sy in count_syllables.keys():
                count_syllables[sy] += 1
            else:
                count_syllables[sy] = 1
```

```python
len(count_syllables)
```

```python
def filter_freq(count_syllables, threshold=10):
    selected_syllables = []
    for k, v in count_syllables.items():
        if v > threshold:
            selected_syllables.append(k)
    print(f"{len(count_syllables)} -> {len(selected_syllables)} with min={threshold}")
    return selected_syllables
```

```python
selected_syllables = filter_freq(count_syllables, threshold=10)
```
